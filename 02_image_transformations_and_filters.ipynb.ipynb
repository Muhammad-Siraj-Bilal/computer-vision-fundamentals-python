```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('steve.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
flipped = cv2.flip(image,0)
plt.imshow(flipped)
```

```python
flipped = cv2.flip(image,1)
plt.imshow(flipped)
```

```python
flipped = cv2.flip(image,-1)
plt.imshow(flipped)
```

```python
rotated = cv2.rotate(image,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotated)
```

```python
rotated = cv2.rotate(image,cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(rotated)
```

```python
rotated = cv2.rotate(image,cv2.ROTATE_180)
plt.imshow(rotated)
```

```python
identity_kernel = np.array([[0,0,0],[0,1,0],[0,0,0]])
identity_kernel
```

```python
identity_image = cv2.filter2D(src=image,ddepth=-1,kernel=identity_kernel)
plt.imshow(identity_image)
```

```python
average_kernel = np.array([[1,1,1],[1,1,1],[1,1,1]])/9
average_image = cv2.filter2D(src=image,ddepth=-1,kernel=average_kernel)
plt.imshow(average_image)
```

```python
v1 = np.array([[1],[4],[6],[4],[1]])
v2 = v1.T
gaussian_kernel = np.dot(v1,v2)/np.sum(gaussian_kernel)
gaussian_kernel
```

```python
gaussian_image = cv2.filter2D(src=image,ddepth=-1,kernel=gaussian_kernel)
plt.imshow(gaussian_image)
```

```python
average_image = cv2.blur(image,(3,3))
plt.imshow(average_image)
```

```python
guassian_image = cv2.GaussianBlur(image,(7,7),5)
```

```python
plt.imshow(guassian_image)
```

```python
median_image = cv2.medianBlur(image,11)
plt.imshow(median_image)
```

```python
edge_kernel = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]]) + np.array([[0,0,0],[0,1,0],[0,0,0]])
edge_image = cv2.filter2D(src=image,ddepth=-1,kernel=edge_kernel)
plt.imshow(edge_image)
```

```python
image_for_canny = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
canny_edge = cv2.Canny(image_for_canny,100,200)
plt.imshow(canny_edge,cmap='gray')
```
