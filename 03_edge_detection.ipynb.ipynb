```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
image = cv2.imread('black_circle.jpg',cv2.IMREAD_GRAYSCALE)
image = image/255
plt.imshow(image,cmap='gray')
```

#### First Derivative

```python
kx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]]) # Prewitts
kx = np.array([[-1,0,1],[-1,0,1],[-1,0,1]]) # Sobel
ky = np.rot90(kx)
ky
```

```python
Gx = ndimage.convolve(image,kx)
Gy = ndimage.convolve(image,ky)
```

```python
plt.imshow(Gx,cmap='gray')
```

```python
plt.imshow(Gy,cmap='gray')
```

```python
# Gm = np.hypot(Gy,Gx)
Gm = np.sqrt(np.square(Gx)+np.square(Gy))
plt.imshow(Gm,cmap='gray')
```

```python
orientation = np.arctan2(Gy,Gx)
plt.imshow(orientation,cmap='gray')
```
