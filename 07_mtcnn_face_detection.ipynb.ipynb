```python
from mtcnn.mtcnn import MTCNN
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('group.webp')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
model = MTCNN()
print(model)
```

```python
output = model.detect_faces(image)
output
```

```python
face_count = len(output)
face_count
```

```python
for i in range(0,face_count):
    box = output[i]['box']
    confidence = output[i]['confidence']
    print(box,confidence)
    if confidence > 0.6:
        cv2.rectangle(image,(box[0],box[1]),(box[0]+box[2],box[1]+box[3]),(0,255,0),5)
plt.imshow(image)
```
